```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
```

```python
def pdf_fun(x):
    y = 1/8 + 3 * x/8
    return y
```

```python
pdf_fun(8)
```

```python
x_values = np.linspace(0, 2, 10000)
y_values = pdf_fun(x_values)
```

```python
plt.figure(figsize = (10, 7))
plt.plot(x_values, y_values)
```

```python
def cdf_fun(x):
    return x/8 + 3/16 * (x **2)
```

```python
y_values_cdf = cdf_fun(x_values)
```

```python
plt.figure(figsize = (10, 7))
plt.plot(x_values, y_values_cdf)
```

```python
mean = 100
stdev = 15
variance = stdev ** 2
```

```python
st.norm.pdf(0)
```

```python
st.norm.pdf(0, loc = mean, scale = stdev)
```

```python
x = np.linspace(-4, 4, 100000)
```

```python
y_norm = st.norm.pdf(x)
```

```python
y_norm_cdf = st.norm.cdf(x)
```

```python
st.norm.cdf(mean + stdev, loc = mean, scale = stdev) - st.norm.cdf(mean-stdev, loc = mean, scale = stdev)
# - st.norm.cdf(mean-stdev)
```

```python
y_norm_isf = st.norm.ppf(x)
```

```python
plt.figure(figsize = (15, 8))
plt.plot(x, y_norm)
plt.plot(x, y_norm_cdf)
plt.show()

plt.figure(figsize = (15, 8))
plt.plot(x, y_norm_isf)
plt.show()
```
